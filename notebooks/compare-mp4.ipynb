```python
import cv2

def compare_mp4_files(file1, file2):
    # Open the video files
    video1 = cv2.VideoCapture(file1)
    video2 = cv2.VideoCapture(file2)

    # Get video properties
    width1 = int(video1.get(cv2.CAP_PROP_FRAME_WIDTH))
    height1 = int(video1.get(cv2.CAP_PROP_FRAME_HEIGHT))
    duration1 = video1.get(cv2.CAP_PROP_POS_MSEC) / 1000.0
    frame_count1 = int(video1.get(cv2.CAP_PROP_FRAME_COUNT))

    width2 = int(video2.get(cv2.CAP_PROP_FRAME_WIDTH))
    height2 = int(video2.get(cv2.CAP_PROP_FRAME_HEIGHT))
    duration2 = video2.get(cv2.CAP_PROP_POS_MSEC) / 1000.0
    frame_count2 = int(video2.get(cv2.CAP_PROP_FRAME_COUNT))

    # Compare the properties
    if width1 == width2:
        print("Width: The two videos have the same width.")
    else:
        print("Width: The two videos have different widths.")

    if height1 == height2:
        print("Height: The two videos have the same height.")
    else:
        print("Height: The two videos have different heights.")

    if duration1 == duration2:
        print("Duration: The two videos have the same duration.")
    else:
        print("Duration: The two videos have different durations.")

    if frame_count1 == frame_count2:
        print("Frame Count: The two videos have the same number of frames.")
    else:
        print("Frame Count: The two videos have different numbers of frames.")

    print(f"Frame Count (Video 1): {frame_count1}")
    print(f"Frame Count (Video 2): {frame_count2}")

    # Release the video files
    video1.release()
    video2.release()

# Usage example
file1 = "/tf/Daniel/custom_lipread_mp4/AGAIN/train/AGAIN_00001.mp4"
file2 = "/tf/Daniel/converted_mp4s/AGAIN/train/AGAIN_01010.mp4"
compare_mp4_files(file1, file2)
```

```python
import cv2
import os
import numpy as np
import shutil

def convert_mp4_files(source_dir, target_dir):
    # Create the target directory if it doesn't exist
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)

    # Specify the list of .mp4 files to convert
    mp4_files = [
        "MOVING_01001.mp4",
        "MOVING_01002.mp4",
        "MOVING_01003.mp4",
        "MOVING_01004.mp4",
        "MOVING_01005.mp4",
        "MOVING_01006.mp4",
        "MOVING_01007.mp4",
        "MOVING_01008.mp4",
        "MOVING_01009.mp4",
        "MOVING_01010.mp4",
        "MOVING_01011.mp4",
        "MOVING_01012.mp4",
        "MOVING_01013.mp4",
        "MOVING_01014.mp4",
        "MOVING_01015.mp4",
        "MOVING_01016.mp4",
        "MOVING_01017.mp4"
    ]

    for file in mp4_files:
        # Get the file path
        file_path = os.path.join(source_dir, file)

        # Open the video file
        video = cv2.VideoCapture(file_path)

        # Get video properties
        fps = video.get(cv2.CAP_PROP_FPS)
        frame_width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
        frame_height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
        total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))

        # Create a directory for the converted frames
        frame_dir = os.path.join(target_dir, os.path.splitext(file)[0])
        os.makedirs(frame_dir, exist_ok=True)

        # Calculate the frame indices to extract 29 frames evenly
        frame_indices = [int(idx) for idx in np.linspace(0, total_frames - 1, 29)]

        # Extract and resize frames
        frames = []
        for frame_index in frame_indices:
            # Set the frame index
            video.set(cv2.CAP_PROP_POS_FRAMES, frame_index)

            # Read the frame
            ret, frame = video.read()

            # Resize the frame to 256x256 pixels
            frame = cv2.resize(frame, (256, 256))

            # Append the resized frame to the list
            frames.append(frame)

        # Release the video file
        video.release()

        # Create the output video file path
        output_file = os.path.join(target_dir, f"{os.path.splitext(file)[0]}.mp4")

        # Write the resized frames to a new video file
        out = cv2.VideoWriter(output_file, cv2.VideoWriter_fourcc(*"mp4v"), fps, (256, 256))
        for frame in frames:
            out.write(frame)
        out.release()

        # Delete the directory containing the frames
        shutil.rmtree(frame_dir)

# Usage example
source_dir = "/tf/Daniel/custom_lipread_mp4/MOVING/train"
target_dir = "/tf/Daniel/converted_mp4s/MOVING/train/"

convert_mp4_files(source_dir, target_dir)
```
