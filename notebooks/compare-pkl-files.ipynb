```python
import pickle
import torch
import numpy as np
from torchvision.io import read_video

def persistent_load(saved_id):
    try:
        return pickle._getattribute(saved_id)
    except AttributeError:
        return None

def compare_pickle_files(file1, file2):
    with open(file1, 'rb') as f1, open(file2, 'rb') as f2:
        data1 = torch.load(f1, map_location=torch.device('cpu'))
        data2 = torch.load(f2, map_location=torch.device('cpu'))

        print(data1['duration'])
        print(data2['duration'])
    differences_found = False

    if not np.array_equal(data1['video'], data2['video']):
        print("Difference found in 'video' field")
        differences_found = True

    if not np.array_equal(data1['label'], data2['label']):
        print("Difference found in 'label' field")
        differences_found = True

    if np.any(data1['duration'] != data2['duration']):
        print("Difference found in 'duration' field")
        differences_found = True

   # Compare the video frames
    video_paths1 = data1['video']
    video_paths2 = data2['video']

    for path1, path2 in zip(video_paths1, video_paths2):
        video1, _, _ = read_video(path1)
        video2, _, _ = read_video(path2)

        if not np.array_equal(video1, video2):
            print("Difference found in 'video' field")
            differences_found = True
            break

    if not np.array_equal(video1, video2):
        print("Difference found in 'video' field")
        differences_found = True


    if not differences_found:
        print("No differences found between the files.")

    # Additional logic to handle the Boolean arrays in the 'duration' field
    if np.any(data1['duration']) and np.any(data2['duration']):
        duration_diff = np.logical_xor(data1['duration'], data2['duration'])
        indices = np.where(duration_diff)[0]
        print("Different durations found at indices:", indices)

# Provide the paths of the .pkl files to compare
file_path1 = '/tf/loqui/custom_lipread_pkls/ACTION/train/ACTION_01018.pkl'
file_path2 = '/tf/loqui/custom_lipread_pkls/ACTION/train/ACTION_00001.pkl'

compare_pickle_files(file_path1, file_path2)
```
